```python
##Creating Child Class --> HourlyEmployee

from Employee import Employee 
class HourlyEmployee(Employee):
    def __init__(self, ename,eadd,hW,hR):
        super().__init__(ename,eadd)
        self.__hourlyWorked=hW
        self.__hourlyRate=hR

##Writing accessor and mutators for variables specific to HourlyEmployee Class

    def set_hourlyWorked(self):
        return self.__HourlyWorked
    
    def get_hourlyWorked(self):
        self.___HourlyWorked=hW
    
    def set_hourlyRate(self):
        return self.__HourlyWorked
    
    def get_hourlyRate(self):
        self.___HourlyWorked=hR
    
    def compute_compensation(self):
        rate=self.get_HourlyWorked()
        if rate>40:
            return (40*self.__hourlyRate)+((self.__hourlyWorked-40)*self.__hourlyRate*1.8)
    
 


        
```

```python
class Vehicle():

    ##Creating instance variables:
    
    def __init__(self,make,model,ym,ml):
        self.__make=make
        self.__model=model
        self.__yearofmanufacture=ym
        self.__mileage=ml
    
    def get_make(self):
        return self.__make
    
    def set_make(self):
        self.___make=make
    
    def get_model(self):
        return self.__model
    
    def set_model(self):
        self.___model=model
    
    def get_yearofmanufacture(self):
        return self.__yearofmanufacture
    
    def set_yearofmanufacture(self):
        self.___yearofmanufacture=ym
    
    def get_mileage(self):
        return self.__mileage
    
    def set_mileage(self):
        self.___mileage=ml 
    
    def __str__():
        return "Vehicle's make"+self.__make +";Vehicle's model"+self.__model+ ";Year of Manufacture" +self.__yearofmanufacture+";Mileage"+self.__mileage
```
